```python
# Your code goes here

# import pandas as pd
# import numpy as np
# read data from 3  url  and combine them into a single dataframe
# url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv
# url2 = https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
# url3 = https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

import pandas as pd
import numpy as np
url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
```

```python
# read url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
# into a dataframe and print the first 5 rows
df1 = pd.read_csv(url1)
print(df1.head())

# display dataframe shape
print("Shape of df1:", df1.shape)
```

```python
# read url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
df2 = pd.read_csv(url2)
print(df2.head())

# Display dataframe shape
print("Shape of df2:", df2.shape)
```

```python
# read url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df3 = pd.read_csv(url3)
print(df3.head())

# display dataframe shape
print("Shape of df3:", df3.shape)
```

```python
# Concatenate the three dataframes into a single dataframe
# import pandas as pd
# import numpy as np
# read data from 3  url  and combine them into a single dataframe
# url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv
# url2 = https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
# url3 = https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv


url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
df = pd.concat([df1, df2, df3], ignore_index=True)
# print the first 5 rows of the dataframe
print(df.head())    

# display dataframe shape
print("Shape of combined dataframe:", df.shape)
```

```python
# print columns before merge
print("Columns BEFORE merge:", df.columns.tolist())
```

```python
# Display dataframe unique values for all columns
for col in df.columns:
    print(f"Unique values in column '{col}': {df[col].unique()}")
```

```python
# Combine state
if "ST" in df.columns and "State" in df.columns:
    df["State"] = df["State"].combine_first(df["ST"])
    df.drop(columns=["ST"], inplace=True)

# Combine gender
if "GENDER" in df.columns and "Gender" in df.columns:
    df["Gender"] = df["Gender"].combine_first(df["GENDER"])
    df.drop(columns=["GENDER"], inplace=True)

print(df.head())
print("Shape of final dataframe:", df.shape)
```

```python
# print columns after merge
print("Columns AFTER merge:", df.columns.tolist())
```

```python
# Create a copy of the combined dataframe to preserve the raw data
df_final = df.copy()
# Display the first 5 rows of the final dataframe
print(df_final.head())
print("Shape of final dataframe copy:", df_final.shape)
```

```python
# Standardize column names for consistency:

# 1. Convert to lowercase
# 2. Replace spaces with underscores
df_final.columns = df_final.columns.str.lower().str.replace(' ', '_')
print("Standardized column names:", df_final.columns.tolist())
```

```python
# Check the shape and column names to confirm the dataset structure
print("Final dataframe shape:", df_final.shape)
print("Final dataframe columns:", df_final.columns.tolist())
```

```python
# Identify and handle invalid values in the dataset
# clean these columns with inconsistent and incorrect values to ensure that they only contain valid data
# focus only on cleaning inconsistent values and not involve handling null values (NaN or None).
# standardized Gender column to "M" and "F"
# replaced State abbreviations with its full name or example "AZ": "Arizona", "Cali": "California", "WA": "Washington"
# replace bachelors by Bachelor in education column
# In Customer Lifetime Value, delete the `%` character
# In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"
# ----------------------------------------------------------------------------------------------------------

# standardized Gender column to "M" and "F"
df_final["gender"] = (
    df_final["gender"]
      .astype(str).str.strip().str.upper()
      .replace({"FEMAL": "F", "FEMALE": "F", "MALE": "M"})
      .str[0]
)
df_final.loc[~df_final["gender"].isin(["F","M"]), "gender"] = np.nan


# In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"
df_final['vehicle_class'] = df_final['vehicle_class'].replace({
    'Sports Car': 'Luxury',
    'Luxury SUV': 'Luxury',
    'Luxury Car': 'Luxury'
})
# replaced Bachelors by Bachelor in education column
df_final['education'] = df_final['education'].replace({
    'Bachelors': 'Bachelor'
})

# Format customer_lifetime_value as float
# Remove '%' character and convert to float
# round to 2 decimal places
df_final["customer_lifetime_value"] = (
    df_final["customer_lifetime_value"] 
        .astype(str)
        .str.replace('%', '', regex=False)
        .astype(float)
        .round(2)
)

# replaced state abbreviations with its full name or example "AZ": "Arizona", "Cali": "California", "WA": "Washington"
df_final['state'] = df_final['state'].replace({
    'AZ': 'Arizona',
    'Cali': 'California',
    'WA': 'Washington'
})

# Format number_of_open_complaints as integer
# Extract the middle value from the format '1/x/00'
# convert to string to integer

df_final["number_of_open_complaints"] = (
    df_final["number_of_open_complaints"]
      .astype(str)
      .str.split("/")
      .str[1]
)
df_final["number_of_open_complaints"] = pd.to_numeric(
    df_final["number_of_open_complaints"].astype(str),
    errors="coerce"
).astype("Int64")

# format monthly_premium_auto as integer
df_final["monthly_premium_auto"] = pd.to_numeric(
    df_final["monthly_premium_auto"].astype(str),
    errors="coerce"
).astype("Int64")

# format income as integer
df_final["income"] = pd.to_numeric(
    df_final["income"].astype(str),
    errors="coerce"
).astype("Int64")

# format total_claim_amount as float
# round to 2 decimal places
df_final["total_claim_amount"] = pd.to_numeric(
    df_final["total_claim_amount"].astype(str),
    errors="coerce"
).round(2)

# Display the first 5 rows after cleaning
print(df_final.head())

# Display unique values in each column after cleaning
for col in df_final.columns:
    print(f"\nUnique values in column '{col}': {df_final[col].unique()}")

# Check the shape and column names to confirm the dataset structure
print("\nFinal dataframe shape:", df_final.shape)
print("\nFinal dataframe columns:", df_final.columns.tolist())
```

```python
# prin final dataframe variable types
print("Final dataframe variable types:")
print(df_final.dtypes)
```

```python
# save the cleaned dataframe to a new CSV file
df_final.to_csv("cleaned_data.csv", index=False)
```

```python
# Your code goes here
# Inspect dataframe 
# url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"
url_marketing = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"

marketing_df = pd.read_csv(url_marketing)

marketing_df.head()

marketing_df.info()

# display dataframe shape
print("\nDataframe shape:", marketing_df.shape)

# display dataframe columns
print("\nDataframe columns:", marketing_df.columns.tolist())

# display dataframe variable types
print("\nDataframe variable types:")
print(marketing_df.dtypes)

# display dataframe unique values for each column
for col in marketing_df.columns:
    print(f"\nUnique values in column '{col}': {marketing_df[col].unique()}")
    
# display dataframe statistical summary
print("\nDataframe statistical summary:")
print(marketing_df.describe(include='all'))

```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# # Analyze which sales channel generates the highest total revenue.
# Create a pivot table to summarize total revenue by sales channel.
# channels: branch, call center, web, and mail.
# Round the total revenue to 2 decimal points.  


# Analyze which sales channel generates the highest total revenue
revenue_by_channel = (
    marketing_df.pivot_table(
        values="monthly_premium_auto",
        index="sales_channel",
        aggfunc="sum"
    )
    .round(2)
)


print("\nTotal revenue by sales channel:")
print(revenue_by_channel)

# Identify the top-performing channel
top_channel = revenue_by_channel["monthly_premium_auto"].idxmax()
print(f"\nThe sales channel with the highest total revenue is: {top_channel}")
```

**Insight: Total Revenue by Sales Channel**
>
> The variable monthly_premium_auto is used as a proxy for revenue, as it represents the recurring premium paid by customers.
>
> This is not a complete or exact revenue calculation, since it does not account for:
>    - policy duration
>    - customer churn
>    - cancellations
>    - discounts or refunds
>    - claim payouts
>
> Therefore, the results should be interpreted as an approximation rather than actual total sales revenue.
>
> Additionally, the dataset does not include a "Mail" sales channel, so revenue from this channel cannot be analyzed.

```python
# Create a pivot table that shows the average customer lifetime value per gender and education level. 
# Analyze the resulting table to draw insights.
avg_clv = (
    marketing_df.pivot_table(
        values="customer_lifetime_value",
        index=["gender", "education"],
        aggfunc="mean"          
    )
    .round(2)
)
print("\nAverage Customer Lifetime Value by Gender and Education Level:")
print(avg_clv)
```

**Insights: Average Customer Lifetime Value by Gender and Education Level**
>
> Gender alone does not strongly determine customer lifetime value.
Differences in CLV are more pronounced across education levels than between genders.
>
> Customer Lifetime Value varies more by education level than by gender:
>
> Customers with lower education levels (High School or Below) show relatively high CLV, 
especially among female customers. Among higher education levels, male customers tend to 
slightly outperform females in terms of average CLV.


## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Your code goes here
# Analyze the months with the highest number of complaints by policy type category.
# Create a sumary table showing the number of complaints by policy type and month.
# Show the results in a long format table.
complaints_summary = (
    marketing_df.pivot_table(
        values="number_of_open_complaints",
        index="policy_type",
        columns="month",
        aggfunc="sum",
        fill_value=0
    )
    .reset_index()
    .melt(id_vars="policy_type", var_name="month", value_name="number_of_complaints")
)
print("\nNumber of Complaints by Policy Type and Month:")
print(complaints_summary)
```
